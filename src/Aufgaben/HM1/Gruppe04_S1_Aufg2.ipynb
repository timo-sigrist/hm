```python
import numpy as np
import matplotlib.pyplot as plt

def gruppe04_S1_Aufg2(a, xmin, xmax):
    x = np.arange(xmin, xmax, 0.1)
    vektor = np.array(a)
    shape = np.array(vektor.shape)

    if np.count_nonzero(shape == 1) == 1 or np.count_nonzero(shape == 1) == 2 or shape.size == 1:
        nVar = shape[1] if shape[0] == 1 else shape[0]
        nVar -= 1

        polynom = 0
        n = nVar
        for a in np.nditer(vektor):
            polynom += a * x**n
            n -= 1

        polynomAbleitung = 0
        n = nVar
        for a in np.nditer(vektor):
            polynomAbleitung += n * a * x**(n - 1)
            n -= 1

        polynomSF = 0
        n = nVar
        for a in np.nditer(vektor):
            polynomSF += (a * x ** (n + 1)) / (n + 1)
            n -= 1

        plt.plot(x, polynom)
        plt.plot(x, polynomAbleitung)
        plt.plot(x, polynomSF)
        plt.xlim(xmin, xmax)
        plt.ylim(-10, 10)
        plt.grid()
        plt.legend(["Polynom", "Ableitung", "Stammfunktion"])


    else:
        raise Exception('Fehler')


x1 = np.array([[-1], [0], [1]])
gruppe04_S1_Aufg2(x1, -5, 5)
```
