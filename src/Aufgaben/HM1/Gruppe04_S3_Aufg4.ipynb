## Aufgabe 4a

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1.0,1.2, 0.01)
h = np.sqrt(100*x**2 - 200*x + 99)

plt.plot(x, h)
plt.grid()
plt.xlim(1, 1.2)
plt.ylim(-0.5,2)
```

Die Auslöschung passiert, weil man die Wurzel von 0 zieht. Daher beginnt der Graph erst bei ungefähr 1.12
