```python
import numpy as np
import matplotlib.pyplot as plt

# a)
x = np.array([0, 2, 9, 13])
y = np.array([150, 104, 172, 152])
polynom_dritten_grades = np.polyfit(x, y, 3)

print('polynom_dritten_grades = ' + str(polynom_dritten_grades))

x2 = np.arange(-1, 14, 0.1)
y2 = polynom_dritten_grades[0] * x2 ** 3 + polynom_dritten_grades[1] * x2 ** 2 + polynom_dritten_grades[2] * x2 ** 1 + polynom_dritten_grades[3] * x2 ** 0

plt.title('e Anzahl der Tage pro Jahr mit extremer UV-Belastung in Hawaii')
plt.plot(x2 + 1997, y2)
plt.grid()
plt.xlabel('Zeit')
plt.ylabel('Tage')
plt.legend()
plt.show()

# b)
x3 = 2003 - 1997
x4 = 2004 - 1997
schaetzwerte2003 = np.polyval(polynom_dritten_grades, x3)
schaetzwerte2004 = np.polyval(polynom_dritten_grades, x4)

print('schaetzwerte2003 = ' + str(schaetzwerte2003))
print('schaetzwerte2004 = ' + str(schaetzwerte2004))

# c)
# Da wir bereits mit der polyfit funktion gearbeitet haben verstehen wir die Aufgabe nicht.
```
