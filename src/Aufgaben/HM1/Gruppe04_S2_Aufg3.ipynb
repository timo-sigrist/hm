```python
import numpy as np
import matplotlib.pyplot as plt

durchmesser = 2.0
seitenzahl = 6
s6 = 1.0
max = 100000000
xWerte = np.array([])
yWerte = np.array([])

def calcUmfangA(seitenanzahl, vorgehendeFolge, xWerte, yWerte):
    seitenanzahl = seitenanzahl*2
    neueFolge = np.sqrt(durchmesser-durchmesser * np.sqrt(1-(np.square(vorgehendeFolge)/4)))
    umfang = seitenanzahl * neueFolge
    print("A - Umfang: " + str(seitenanzahl) + " = " + str(umfang))
    print("A - Differenz zum eigentlichen pi:" +  str(durchmesser * np.pi - seitenanzahl * (neueFolge/2)))
    xWerte = np.append(xWerte, [seitenanzahl])
    yWerte = np.append(yWerte, [umfang])
    if (seitenanzahl * 2 < max):
        calcUmfangA(seitenanzahl, neueFolge, xWerte, yWerte)
    else:
        plt.plot(xWerte, yWerte)

def calcUmfangB(seitenanzahl, vorgehendeFolge, xWerte, yWerte):
    seitenanzahl = seitenanzahl*2
    neueFolge = np.sqrt(np.square(vorgehendeFolge)/(2*(1+np.sqrt(1 - np.square(vorgehendeFolge)/4))))
    umfang = seitenanzahl * neueFolge
    print("B - Umfang: " + str(seitenanzahl) + " = " + str(umfang))
    print("B - Differenz zum eigentlichen pi:" +  str(durchmesser * np.pi - seitenanzahl * (neueFolge/2)))
    xWerte = np.append(xWerte, [seitenanzahl])
    yWerte = np.append(yWerte, [umfang])
    if (seitenanzahl * 2 < max):
        calcUmfangB(seitenanzahl, neueFolge, xWerte, yWerte)
    else:
        plt.plot(xWerte, yWerte)


print ("Aufgabe 3a:")
calcUmfangA(seitenzahl, s6, xWerte, yWerte)
print("Es nähert sich immer mehr der Zahl Pi")
print()
print ("Aufgabe 3b:")
calcUmfangB(seitenzahl, s6, xWerte, yWerte)
plt.legend(["Aufabge a", "Aufgabe b"])
plt.ylabel("2 s")
plt.xlabel("Seitenzahl")

# Unsere Bemerkungen:
# Bei A sehen wir, dass nach einer gewissen grösse von n die Zahl einen starken Aussreisser hat und von Pi stark abweicht.
# Bei B sehen wir, dass es sich immer konstant an Pi annähert, auch bei einem grossen n.
```
