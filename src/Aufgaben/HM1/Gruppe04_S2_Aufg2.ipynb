```python
import numpy as np
import matplotlib.pyplot as plt

# a)
x = np.arange(1.99,2.01,0.00004)
f1 = x**7 - 14*x**6 + 84*x**5 - 280*x**4 + 560*x**3 - 672*x**2 + 448*x - 128
f2 = (x - 2)**7

plt.plot(x, f1)
plt.plot(x, f2)

# Was stellen Sie fest? Analytisch sind die
# beiden Polynome identisch aber weshalb weichen die beiden von Python gezeichneten Polynome voneinander ab?
# Weil beim Polynom die Gefahr besteht, dass sich Terme auslöschen (-280x^4+560x^3), bzw. kann es zu Fehler kommen.
# Und weil das Polynom(f1) schlecht kondizioniert ist, weil es zu Rundungsfehlern führen kann.
```

```python
# b)
x = np.arange(-10**-14,10**-14,10**-17)
g = x / (np.sin(1 + x) - np.sin(1))

plt.plot(x, g)

# Ist die numerische Berechnung des
# Grenzwertes limx!0 g(x) auf diese Weise stabil?
# Nein, weil kein Nullpunkt definiert ist.
```

```python
# c)
x = np.arange(-10**-14,10**-14,10**-17)
nenner1 = np.sin(1+x) - np.sin(1)
nenner2 = 2 * np.cos((1+x+1)/2) * np.sin(x/2)
g_new = x/(nenner2)

plt.plot(x, g_new)

# Was stellen Sie fest und weshalb? Welcher Wert nimmt limx!0 g(x) also an?
# Weil es beim Nullpunkt ebenfalls definiert ist, sieht es stabil aus.
```
