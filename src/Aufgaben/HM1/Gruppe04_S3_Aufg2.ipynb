```python
import numpy as np
import matplotlib.pyplot as plt

n = 100000
x = np.arange(0, 9, 1)
potenzieren = x**n
kondPot = np.full((x.size, 1),n ) # da aufgelöst einfach n

wurzelziehen = x**(1/n)
wurzelziehenAbleitung  = (x**((1/n)-1))/n
kondWurzel = (wurzelziehenAbleitung * x) / wurzelziehen

plt.plot(x, kondPot)
```

Umso grösser n wird umso grösser wird auch die Konditionszahl. Daher ist es ein schlecht Konditioniertes Problem.

```python
plt.plot(x, kondWurzel)
```

Die Konditionszahl nähert sich mit einem grossen n immer mehr dem 0 Wert. Daher ist es ein gut Konditioniertes Problem.
