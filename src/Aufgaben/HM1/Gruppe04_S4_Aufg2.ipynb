# Aufgabe 2a

```python
import numpy as np
import matplotlib.pyplot as plt

def loopKrankheit(a,iterationen):
    values = np.array([])
    k = 0.1
    for i in iterationen:
        k = a * k * (1 - k)
        values = np.append(values, [k])
    return values

time = np.arange(1,30,1)

kVals0 = loopKrankheit(0, time)
kVals1 = loopKrankheit(1, time)
kVals2 = loopKrankheit(2, time)
kVals3 = loopKrankheit(3, time)
kVals4 = loopKrankheit(4, time)

print(kVals2)

plt.plot(time, kVals0)
plt.plot(time, kVals1)
plt.plot(time, kVals2)
plt.plot(time, kVals3)
plt.plot(time, kVals4)
plt.xlabel("Zeit")
plt.ylabel("Kranke Kinder")
plt.grid()
plt.legend(["a = 0", "a = 1", "a = 2", "a = 3", "a = 4"])

# Aus der Grafik erkennt man, dass fr Werte  a element in [0,2] auf Fixpunkte zusteuert.
# 3 und 4 steuern auf keine Fixpunkte mehr zu
```

```python
kVals21 = loopKrankheit(2.6, time)
kVals22 = loopKrankheit(2.8, time)
kVals23 = loopKrankheit(3.0, time)
kVals24 = loopKrankheit(3.2, time)
kVals25 = loopKrankheit(3.4, time)

plt.plot(time, kVals21)
plt.plot(time, kVals22)
plt.plot(time, kVals23)
plt.plot(time, kVals24)
plt.plot(time, kVals25)
plt.xlabel("Zeit")
plt.ylabel("Kranke Kinder")
plt.grid()
plt.legend(["a = 2.6", "a = 2.8", "a = 3.0", "a = 3.2", "a = 3.4"])

# Die Grenze liegt ungefhr bei der Steigung a = 3
```
