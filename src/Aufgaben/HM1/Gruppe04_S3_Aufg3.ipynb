## Aufgabe 3a

```python
import numpy as np
import matplotlib.pyplot as plt

c = 2
a = 3
x = np.arange(0, 5, 1)

exp = c * a**x

pot = c * x **a


plt.plot(x, np.log (exp))
plt.plot(np.log(x), np.log(pot))
```

```python
x = np.arange(1,100)
f = 5/((2*x**2)**(1/3))
fAbleit = -((5*2)**(2/3))/(3*x**(5/3))


plt.plot(np.log(x), np.log(f))
plt.plot(np.log(x), fAbleit)
plt.plot(np.log(x), f)
```

```python
x = np.arange(1,100)
g = 10**5 * ((2*np.e)**(-x/100))
gAbleit = -125 * np.log(2)*2**(3-x/100) * np.e**(-x/100) - 125 * 2**(3-x/100) * np.e**(-x/100)
plt.plot(x, np.log(g))
plt.plot(x, gAbleit)
plt.plot(x, g)
```

```python
x = np.arange(1,5, 0.1)

h = (10**(2*x)/2**(5*x))**2
hAbleit = np.log(10) * 2**(2-10*x) * 10**(4*x) - 5*np.log(2) * 2**(2-10*(x)) * 10 ** (4*x).astype(int)
plt.plot(x, np.log(h))
plt.plot(x, hAbleit)
```
